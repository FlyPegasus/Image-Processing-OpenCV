```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "E://Development//Image Processing//basic codes//image data//image2.jpg"
img = cv2.imread(path)
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

# Thresholding
- segments image into two regions: foreground and background
- replacing pixels with high or low depending on threshold
- it is performed on gray scale image

```python
# Converting to gray scale
gray_img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Performing Gaussian blur
gaussian_blur_img = cv2.GaussianBlur(gray_img, (11, 11), 0)
plt.imshow(gaussian_blur_img, cmap="gray")
```

## 1. Simple Thresholding

```python
th1, th_img1 = cv2.threshold(gaussian_blur_img, 150, 255, cv2.THRESH_BINARY)
plt.subplot(1, 2, 1)
plt.imshow(th_img1, cmap="gray")
th2, th_img2 = cv2.threshold(gaussian_blur_img, 150, 255, cv2.THRESH_BINARY_INV)
plt.subplot(1, 2, 2)
plt.imshow(th_img2, cmap="gray")
plt.show()
```

```python
print("Simple threshold value used: ", th1)
```

## 2. OTSU Thresholding

```python
th3, th_img3 = cv2.threshold(gaussian_blur_img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
plt.imshow(th_img3, cmap="gray")
```

```python
print("OTSU threshold value used: ", th3)
```

## 3. Adaptive Thresholding

```python
th_img4 = cv2.adaptiveThreshold(gaussian_blur_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
#                                                        mean kernel                             11: kernel size 2: tuning
plt.imshow(th_img4, cmap="gray")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Reading Image

```python
path = "E://Development//Image Processing//basic codes//image data//image1.jpg"
img1 = cv2.imread(path)
type(img1)
```

```python
plt.subplot(1, 3, 1)
plt.imshow(img1) # gives inverted color

# So to get correct image
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.subplot(1, 3, 2)
plt.imshow(img2)


# convertion to gray image
img3 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
plt.subplot(1, 3, 3)
plt.imshow(img3, cmap="gray")
plt.show()
```

# Splitting and Merging of Channels

```python
b, g, r = cv2.split(img1)
plt.subplot(1, 3, 1)
plt.imshow(b, cmap="gray")
plt.title("Blue")
plt.subplot(1, 3, 2)
plt.imshow(g, cmap="gray")
plt.title("Green")
plt.subplot(1, 3, 3)
plt.imshow(r, cmap="gray")
plt.title("Red")
```

```python
merged_img = cv2.merge((b, g, r))
plt.imshow(merged_img)
```

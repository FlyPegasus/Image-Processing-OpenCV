```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "E://Development//Image Processing//basic codes//image data//image1.jpg"
img = cv2.imread(path)
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Types of Smoothing
## 1. Mean Blurring

```python
mean_blur_img = cv2.blur(img1, (121, 121))
plt.imshow(mean_blur_img)
```

## 2. Median Blur

```python
median_blur_img = cv2.medianBlur(img1,51)
plt.imshow(median_blur_img)
```

## 3. Gaussian Blur

```python
gaussian_blur_img = cv2.GaussianBlur(img1, (11, 11), 0)
plt.imshow(gaussian_blur_img)
```

## 4. Bilateral Blur

# Edge Detection
- locates rapid change in intensity
- useful in capturing visual information
### Three types of Edges
- step edge
- ramp edge
- roof edge

## 1. Using Gradients
- catching rate of change of intensity
- strength of orientation

```python
# Sobel X-axis
plt.subplot(2, 2, 1)
img3 = cv2.Sobel(img2,cv2.CV_64F,dx=1,dy=0,ksize=1)
plt.imshow(img3, cmap="gray")
# Sobel Y-axis
plt.subplot(2, 2, 2)
img4 = cv2.Sobel(img2,cv2.CV_64F,dx=0,dy=1,ksize=1)
plt.imshow(img4, cmap="gray")
plt.subplot(2, 2, 3)
img5 = cv2.Sobel(img2,cv2.CV_64F,dx=1,dy=1,ksize=1)
plt.imshow(img5, cmap="gray")
plt.show()
```

## 3. Using Canny Edge
1. Noise reduction
2. Intensity gradient of image
3. Suppression of false edges
4. Hysteresis Thresholding

Hysteresis Thresholding has two values for thresholding (t1, t2) where t1 is the higher threshold and t2 is the lower threshold, the edges having lower intensity than t2 are rejected.

```python
cannyEdgeImg = cv2.Canny(imag2, threshold1 = 100, threshold2 = 200)
plt.imshow(cannyEdgeImg, cmap="gray")
plt.title("Canny Edge Detection")
```
